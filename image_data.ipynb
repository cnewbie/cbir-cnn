```python
import os
import time
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16.0, 16.0)
plt.rcParams['image.interpolation'] = 'nearest'


%load_ext autoreload
%autoreload 2
```

```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, load_img, img_to_array
from cbir.utils import get_list

img_input_path = '/home/jy/dataset/fabric_dataset/new'
img_out_path = '/home/jy/dataset/fabric_dataset/out'

img_list = get_list(img_input_path)
img_num = len(img_list)
print(img_num)


datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.2,
    height_shift_range=0.3,
    horizontal_flip=True,
    zoom_range=0.2,)
img_names=list()
for i,img in enumerate(img_list):
    if i != 0 and i % 100 == 0:
        print('process %.2f ' %(i/img_num,))
    img_name = os.path.basename(img).split('.')[0]
    img_names.append(img_name)
    print(img_name)
    img_data = np.expand_dims(img_to_array(load_img(img)),axis=0)
    #print(img_data.shape)
    k = 0
    for j in datagen.flow(
                        img_data,
                        batch_size=1,
                        save_to_dir=img_out_path,
                        save_prefix=img_name,
                        save_format='jpg'):
        #print(j[0].shape,j[1])
        k = k + 1
        if k > 5:
            break
```

```python
out_imgs = get_list(img_out_path)
gt_file = '/home/jy/dataset/fabric_dataset/gt_new.txt'
tmp_dict = dict()
out_dict = dict()
k = 1
for i in out_imgs:
    tmp = os.path.basename(i).split('_')[0]
    if tmp_dict.get(tmp) is None:
        pass
        tmp_dict[tmp] = k
        out_dict[tmp] = k
        k += 1
    out_dict[os.path.basename(i).split('.')[0]] = tmp_dict.get(tmp)

print(len(out_dict))
with open(gt_file, 'w') as f:
    f.writelines([k+' '+str(v)+ '\n' for k,v in sorted(out_dict.items())])
for k,v in sorted(out_dict.items()):
    print(k,v)
    if v > 10:
        break


#clean memory
del k
del out_imgs
del tmp_dict
del gt_file
```
